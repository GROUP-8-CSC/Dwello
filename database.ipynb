```python
import os
import psycopg2
from dotenv import load_dotenv

# 1. Load the environment variables from the .env file
load_dotenv()

def connect_to_db():
    connection = None
    try:
        # 2. Fetch the connection string from the environment
        db_url = os.getenv("DATABASE_URL")
        
        # 3. Connect to the PostgreSQL database
        print("Connecting to the PostgreSQL database...")
        connection = psycopg2.connect(db_url)
        
        # 4. Create a cursor to execute SQL commands
        cursor = connection.cursor()
        
        # 5. Execute a test query to verify it works
        cursor.execute("SELECT version();")
        db_version = cursor.fetchone()
        
        print("\n🎉 Connection Successful!")
        print(f"PostgreSQL version: {db_version[0]}\n")
        
        # Close the cursor
        cursor.close()

    except (Exception, psycopg2.DatabaseError) as error:
        print(f"❌ Error while connecting to PostgreSQL: {error}")
        
    finally:
        # 6. Always ensure the connection closes when done
        if connection is not None:
            connection.close()
            print("Database connection closed.")

if __name__ == "__main__":
    connect_to_db()
```
