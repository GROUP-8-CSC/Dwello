```python
import tkinter as tk
from tkinter import ttk, font
import math


# ── Colour palette ──────────────────────────────────────────────────────────
BG_WHITE       = "#FFFFFF"
BG_DARK        = "#0D1B0F"          # very dark green-black (hero side)
GREEN_PRIMARY  = "#1A7A2E"          # Lagos-green CTA
GREEN_LIGHT    = "#25A244"          # hover / accent
GREEN_TEXT     = "#1A7A2E"
GOLD_ACCENT    = "#C8A84B"          # warm gold for tagline text
TEXT_DARK      = "#1A1A1A"
TEXT_MID       = "#555555"
TEXT_LIGHT     = "#AAAAAA"
BORDER_COLOR   = "#E0E0E0"
INPUT_BG       = "#F7F9F7"
TAB_ACTIVE_BG  = "#FFFFFF"
TAB_BAR_BG     = "#F0F4F0"


class LagosHomeFinderApp(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title("Lagos Home Finder")
        self.geometry("780x520")
        self.resizable(False, False)
        self.configure(bg=BG_WHITE)

        # Remove default window chrome and use custom
        self._build_titlebar()
        self._build_main()

    # ── Custom title bar ─────────────────────────────────────────────────────
    def _build_titlebar(self):
        bar = tk.Frame(self, bg="#1C1C1C", height=30)
        bar.pack(fill="x", side="top")
        bar.pack_propagate(False)

        # Window controls (right side)
        ctrl = tk.Frame(bar, bg="#1C1C1C")
        ctrl.pack(side="right", padx=10)

        for symbol, cmd, hover in [
            ("×", self.destroy,  "#E74C3C"),
            ("—", self.iconify,  "#F39C12"),
        ]:
            btn = tk.Label(ctrl, text=symbol, bg="#1C1C1C", fg="#AAAAAA",
                           font=("Arial", 11), cursor="hand2", padx=6)
            btn.pack(side="right")
            btn.bind("<Button-1>", lambda e, c=cmd: c())
            btn.bind("<Enter>",    lambda e, b=btn, h=hover: b.config(fg=h))
            btn.bind("<Leave>",    lambda e, b=btn: b.config(fg="#AAAAAA"))

        # Drag support
        bar.bind("<ButtonPress-1>",   self._start_drag)
        bar.bind("<B1-Motion>",       self._do_drag)

    def _start_drag(self, event):
        self._drag_x = event.x
        self._drag_y = event.y

    def _do_drag(self, event):
        dx = event.x - self._drag_x
        dy = event.y - self._drag_y
        x  = self.winfo_x() + dx
        y  = self.winfo_y() + dy
        self.geometry(f"+{x}+{y}")

    # ── Main two-panel layout ────────────────────────────────────────────────
    def _build_main(self):
        container = tk.Frame(self, bg=BG_WHITE)
        container.pack(fill="both", expand=True)

        self._build_hero_panel(container)
        self._build_form_panel(container)

    # ── LEFT: Hero panel ─────────────────────────────────────────────────────
    def _build_hero_panel(self, parent):
        hero = tk.Canvas(parent, width=295, height=490, bg=BG_DARK,
                         highlightthickness=0)
        hero.pack(side="left", fill="y")

        # Background gradient-like layering with rectangles
        hero.create_rectangle(0, 0, 295, 490, fill="#0D1B0F", outline="")
        # Subtle green glow at the bottom (layered dark-green strips)
        glow_shades = [
            "#0D1F0D", "#0E210E", "#0F230F", "#102510", "#112711",
            "#122912", "#132B13", "#142D14", "#152F15", "#163116",
        ]
        for i, shade in enumerate(glow_shades):
            y0 = 490 - (len(glow_shades) - i) * 12
            hero.create_rectangle(0, y0, 295, y0 + 13, fill=shade, outline="")

        # City silhouette (simple geometric shapes)
        self._draw_cityscape(hero)

        # Brand logo area
        hero.create_oval(22, 30, 46, 54, fill=GREEN_PRIMARY, outline="")
        hero.create_text(34, 42, text="🏠", font=("Arial", 10), fill="white")

        hero.create_text(52, 38, text="LAGOS", anchor="w",
                         font=("Arial", 8, "bold"), fill="white")
        hero.create_text(52, 51, text="HOME FINDER", anchor="w",
                         font=("Arial", 7, "bold"), fill=GOLD_ACCENT)

        # Hero headline
        hero.create_text(28, 210, text="Find Your", anchor="w",
                         font=("Georgia", 22, "bold"), fill="white")
        hero.create_text(28, 242, text="Perfect", anchor="w",
                         font=("Georgia", 22, "bold"), fill="white")
        hero.create_text(28, 275, text="Lagos Home", anchor="w",
                         font=("Georgia", 21, "bold"), fill=GOLD_ACCENT)

        # Tag lines with bullets
        for i, tag in enumerate(["Smart.  Affordable.", "Personalised."]):
            y = 308 + i * 22
            hero.create_oval(28, y + 4, 34, y + 10, fill=GREEN_LIGHT, outline="")
            hero.create_text(40, y + 7, text=tag, anchor="w",
                             font=("Arial", 9), fill="#CCCCCC")

        # Bottom watermark
        hero.create_text(147, 472, text="LAGOS HOME FINDER",
                         font=("Arial", 7, "bold"), fill="#405042",
                         anchor="center")

    def _draw_cityscape(self, canvas):
        """Simple geometric Lagos skyline"""
        buildings = [
            # (x, bottom_y, w, h, fill)
            (10,  420, 22, 80,  "#1A2E1A"),
            (35,  420, 18, 55,  "#152515"),
            (56,  420, 26, 100, "#1F351F"),
            (85,  420, 14, 65,  "#162816"),
            (102, 420, 30, 120, "#224022"),
            (135, 420, 20, 75,  "#1A301A"),
            (158, 420, 16, 90,  "#1C321C"),
            (177, 420, 28, 60,  "#172917"),
            (208, 420, 22, 105, "#204020"),
            (233, 420, 18, 70,  "#1E3A1E"),
            (254, 420, 16, 85,  "#1B2E1B"),
            (273, 420, 22, 55,  "#172517"),
        ]
        for x, by, w, h, fill in buildings:
            canvas.create_rectangle(x, by - h, x + w, by,
                                    fill=fill, outline="")
            # Windows
            for row in range(3):
                for col in range(2):
                    wx = x + 4 + col * (w // 2 - 2)
                    wy = by - h + 8 + row * 14
                    if wy < by - 4:
                        canvas.create_rectangle(wx, wy, wx + 4, wy + 6,
                                                fill="#2A4A2A", outline="")

        # Water at bottom
        canvas.create_rectangle(0, 420, 295, 440, fill="#0A1A14", outline="")
        # Water shimmer
        for i in range(0, 295, 20):
            canvas.create_line(i, 428, i + 10, 428,
                               fill="#1A3A2A", width=1)

    # ── RIGHT: Form panel ────────────────────────────────────────────────────
    def _build_form_panel(self, parent):
        panel = tk.Frame(parent, bg=BG_WHITE)
        panel.pack(side="left", fill="both", expand=True)

        inner = tk.Frame(panel, bg=BG_WHITE)
        inner.place(relx=0.5, rely=0.5, anchor="center")

        # Logo
        logo_row = tk.Frame(inner, bg=BG_WHITE)
        logo_row.pack(pady=(0, 20))

        logo_icon = tk.Label(logo_row, text="🏠", font=("Arial", 22),
                             bg=GREEN_PRIMARY, fg="white",
                             width=2, relief="flat")
        logo_icon.pack(side="left", padx=(0, 8))

        logo_text = tk.Frame(logo_row, bg=BG_WHITE)
        logo_text.pack(side="left")
        tk.Label(logo_text, text="LAGOS", bg=BG_WHITE, fg=TEXT_DARK,
                 font=("Arial", 13, "bold")).pack(anchor="w")
        tk.Label(logo_text, text="HOME FINDER", bg=BG_WHITE, fg=GREEN_PRIMARY,
                 font=("Arial", 13, "bold")).pack(anchor="w")

        # Tab bar
        self._active_tab = tk.StringVar(value="signup")
        tab_frame = tk.Frame(inner, bg=TAB_BAR_BG, bd=0,
                             highlightthickness=1,
                             highlightbackground=BORDER_COLOR)
        tab_frame.pack(fill="x", ipady=4, pady=(0, 22))

        self._tab_signup = self._make_tab(tab_frame, "Sign Up",  "signup")
        self._tab_login  = self._make_tab(tab_frame, "Log In",   "login")
        self._tab_signup.pack(side="left", expand=True, fill="x")
        self._tab_login.pack(side="left",  expand=True, fill="x")
        self._highlight_tab("signup")

        # Form fields container
        self._form_frame = tk.Frame(inner, bg=BG_WHITE)
        self._form_frame.pack(fill="x")

        self._build_signup_fields()

        # Bottom link
        self._bottom_link_var = tk.StringVar(value="Already have an account?")
        self._bottom_frame = tk.Frame(inner, bg=BG_WHITE)
        self._bottom_frame.pack(pady=(14, 0))

        self._bottom_label = tk.Label(self._bottom_frame,
                                      textvariable=self._bottom_link_var,
                                      bg=BG_WHITE, fg=TEXT_MID,
                                      font=("Arial", 9))
        self._bottom_label.pack(side="left")

        self._link_label = tk.Label(self._bottom_frame, text="Log In",
                                    bg=BG_WHITE, fg=GREEN_PRIMARY,
                                    font=("Arial", 9, "underline"),
                                    cursor="hand2")
        self._link_label.pack(side="left", padx=(4, 0))
        self._link_label.bind("<Button-1>", lambda e: self._switch_tab("login"))

    def _make_tab(self, parent, text, tab_id):
        btn = tk.Label(parent, text=text, bg=TAB_BAR_BG, fg=TEXT_MID,
                       font=("Arial", 10), cursor="hand2", pady=6)
        btn.bind("<Button-1>", lambda e: self._switch_tab(tab_id))
        return btn

    def _highlight_tab(self, active):
        if active == "signup":
            self._tab_signup.config(bg=TAB_ACTIVE_BG, fg=GREEN_PRIMARY,
                                    font=("Arial", 10, "bold"),
                                    relief="groove", bd=1)
            self._tab_login.config(bg=TAB_BAR_BG, fg=TEXT_MID,
                                   font=("Arial", 10), relief="flat", bd=0)
        else:
            self._tab_login.config(bg=TAB_ACTIVE_BG, fg=GREEN_PRIMARY,
                                   font=("Arial", 10, "bold"),
                                   relief="groove", bd=1)
            self._tab_signup.config(bg=TAB_BAR_BG, fg=TEXT_MID,
                                    font=("Arial", 10), relief="flat", bd=0)

    def _switch_tab(self, tab_id):
        self._active_tab.set(tab_id)
        self._highlight_tab(tab_id)
        for w in self._form_frame.winfo_children():
            w.destroy()

        if tab_id == "signup":
            self._build_signup_fields()
            self._bottom_link_var.set("Already have an account?")
            self._link_label.config(text="Log In")
            self._link_label.bind("<Button-1>",
                                  lambda e: self._switch_tab("login"))
        else:
            self._build_login_fields()
            self._bottom_link_var.set("Don't have an account?")
            self._link_label.config(text="Sign Up")
            self._link_label.bind("<Button-1>",
                                  lambda e: self._switch_tab("signup"))

    # ── Field helpers ────────────────────────────────────────────────────────
    def _input_field(self, parent, icon, placeholder, show=None):
        wrapper = tk.Frame(parent, bg=INPUT_BG,
                           highlightthickness=1,
                           highlightbackground=BORDER_COLOR)
        wrapper.pack(fill="x", pady=5, ipady=2)

        icon_lbl = tk.Label(wrapper, text=icon, bg=INPUT_BG, fg=TEXT_LIGHT,
                            font=("Arial", 11), padx=8)
        icon_lbl.pack(side="left")

        entry_kwargs = dict(bg=INPUT_BG, fg=TEXT_MID, relief="flat",
                            font=("Arial", 10), bd=0,
                            insertbackground=GREEN_PRIMARY, width=24)
        if show:
            entry_kwargs["show"] = show

        entry = tk.Entry(wrapper, **entry_kwargs)
        entry.pack(side="left", fill="x", expand=True, pady=6)

        # Placeholder behaviour
        entry.insert(0, placeholder)
        entry.config(fg=TEXT_LIGHT)

        def on_focus_in(e):
            if entry.get() == placeholder:
                entry.delete(0, "end")
                entry.config(fg=TEXT_DARK)
                if show:
                    entry.config(show=show)

        def on_focus_out(e):
            if not entry.get():
                entry.insert(0, placeholder)
                entry.config(fg=TEXT_LIGHT, show="")

        entry.bind("<FocusIn>",  on_focus_in)
        entry.bind("<FocusOut>", on_focus_out)

        # Focus border highlight
        def on_enter(e):
            wrapper.config(highlightbackground=GREEN_PRIMARY)
        def on_leave(e):
            wrapper.config(highlightbackground=BORDER_COLOR)

        entry.bind("<FocusIn>",  lambda e: (on_focus_in(e), on_enter(e)))
        entry.bind("<FocusOut>", lambda e: (on_focus_out(e), on_leave(e)))

        return entry

    def _cta_button(self, parent, text):
        btn = tk.Label(parent, text=text, bg=GREEN_PRIMARY, fg="white",
                       font=("Arial", 11, "bold"),
                       pady=10, cursor="hand2", relief="flat")
        btn.pack(fill="x", pady=(14, 0))

        btn.bind("<Enter>", lambda e: btn.config(bg=GREEN_LIGHT))
        btn.bind("<Leave>", lambda e: btn.config(bg=GREEN_PRIMARY))
        return btn

    def _build_signup_fields(self):
        self._input_field(self._form_frame, "👤", "Full Name")
        self._input_field(self._form_frame, "✉", "Email Address")
        pw_row = tk.Frame(self._form_frame, bg=INPUT_BG,
                          highlightthickness=1,
                          highlightbackground=BORDER_COLOR)
        pw_row.pack(fill="x", pady=5, ipady=2)

        tk.Label(pw_row, text="🔒", bg=INPUT_BG, fg=TEXT_LIGHT,
                 font=("Arial", 11), padx=8).pack(side="left")

        self._pw_entry = tk.Entry(pw_row, bg=INPUT_BG, fg=TEXT_LIGHT,
                                  relief="flat", font=("Arial", 10), bd=0,
                                  insertbackground=GREEN_PRIMARY, width=20,
                                  show="")
        self._pw_entry.insert(0, "Password")
        self._pw_entry.pack(side="left", fill="x", expand=True, pady=6)

        # Eye toggle
        self._pw_visible = False
        self._eye_btn = tk.Label(pw_row, text="👁", bg=INPUT_BG, fg=TEXT_LIGHT,
                                 font=("Arial", 10), cursor="hand2", padx=8)
        self._eye_btn.pack(side="right")
        self._eye_btn.bind("<Button-1>", self._toggle_pw)

        def pw_focus_in(e):
            if self._pw_entry.get() == "Password":
                self._pw_entry.delete(0, "end")
                self._pw_entry.config(fg=TEXT_DARK, show="•")

        def pw_focus_out(e):
            if not self._pw_entry.get():
                self._pw_entry.config(show="")
                self._pw_entry.insert(0, "Password")
                self._pw_entry.config(fg=TEXT_LIGHT)
            pw_row.config(highlightbackground=BORDER_COLOR)

        self._pw_entry.bind("<FocusIn>",
                            lambda e: (pw_focus_in(e),
                                       pw_row.config(highlightbackground=GREEN_PRIMARY)))
        self._pw_entry.bind("<FocusOut>", pw_focus_out)

        self._cta_button(self._form_frame, "Create Account")

    def _build_login_fields(self):
        self._input_field(self._form_frame, "✉", "Email Address")
        self._input_field(self._form_frame, "🔒", "Password", show="•")

        # Forgot password
        fp = tk.Label(self._form_frame, text="Forgot password?",
                      bg=BG_WHITE, fg=GREEN_PRIMARY,
                      font=("Arial", 8, "underline"), cursor="hand2",
                      anchor="e")
        fp.pack(fill="x", pady=(4, 0))

        self._cta_button(self._form_frame, "Log In")

    def _toggle_pw(self, event):
        self._pw_visible = not self._pw_visible
        if self._pw_visible:
            self._pw_entry.config(show="")
            self._eye_btn.config(fg=GREEN_PRIMARY)
        else:
            if self._pw_entry.get() != "Password":
                self._pw_entry.config(show="•")
            self._eye_btn.config(fg=TEXT_LIGHT)


if __name__ == "__main__":
    app = LagosHomeFinderApp()
    app.mainloop()
```
